###### import necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

###### merge those two csv files (after getting as dataframes, get them as a single dataframe)

```python
df = pd.concat(
    map(pd.read_csv, ['college_1.csv', 'college_2.csv']), ignore_index=True)
```

###### Take each csv file , split that csv file into multiple categories (example csv files are added in the repo) 


###### consider if the codekata score exceeds 15000 points(present week) then make a csv on those observations as Exceeded expectations.csv


###### if  10000<codekata score<15000   (Reached_expectations.csv)



###### if  7000<codekata score<10000   (Needs_Improvement.csv)


###### if  codekate score < 7000        (Unsatisfactory.csv)

```python
exceeded_expectations = df[df['CodeKata Score'] >= 15000]
exceeded_expectations.to_csv('Exceeded expectations.csv', index=False)

reached_expectations = df[(df['CodeKata Score'] >= 10000) & (df['CodeKata Score'] < 15000)]
reached_expectations.to_csv('Reached expectations.csv', index=False)

needs_improvement = df[(df['CodeKata Score'] >= 7000) & (df['CodeKata Score'] < 10000)]
needs_improvement.to_csv('Needs Improvement.csv', index=False)

unsatisfactory = df[df['CodeKata Score'] < 7000]
unsatisfactory.to_csv('Unsatisfactory.csv', index=False)
```

###### Average of previous week geekions vs this week geekions (i.e Previous Geekions vs CodeKata Score)

```python
df['prev_week_geekions_vs_codekata_score'] = df['Previous Geekions'] / df['CodeKata Score']
average = df['prev_week_geekions_vs_codekata_score'].mean()
print(f'Average of previous week geekions vs this week geekions: {average}')
```

###### No of students participated 

```python
num_students = df.shape[0]
print(f'Number of students participated: {num_students}')
```

###### #Average completion of python course or my_sql or python english or computational thinking

```python
courses = ['python', 'mysql', 'python_en', 'computational_thinking']
average_completion = df[courses].mean().mean()
print(f'Average completion of python course or my_sql or python english or computational thinking: {average_completion}')
```

###### rising star of the week (top 3 candidate who performed well in that particular week)

```python
df = df.sort_values(by='CodeKata Score', ascending=False)
rising_star_of_the_week = df.head(3)
print(f'Rising star of the week: {rising_star_of_the_week}')
```

###### Shining stars of the week (top 3 candidates who has highest geekions)

```python
df_top_geekions = df.sort_values(by='Previous Geekions', ascending=False).head(3)
print("Shining stars of the week (top 3 candidates who has highest geekions):")
print(df_top_geekions[['Name', 'Previous Geekions']])

```

###### Department wise codekata performence (pie chart)

```python
df['Department'].value_counts().plot.pie(autopct='%.1f%%')
plt.title('Department wise codekata performence')
plt.show()
```

###### Department wise toppers (horizantal bar graph or any visual representations of your choice)

```python
df_top = df.sort_values(by='CodeKata Score', ascending=False)
df_top = df_top.groupby('Department').head(3)

df_top.plot.barh(x='Department', y='CodeKata Score', title='Department and Student Name')
plt.xlabel('Codekata Score')
plt.ylabel('Department and Student Name')
plt.show()
```
